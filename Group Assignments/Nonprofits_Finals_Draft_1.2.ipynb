```python
import plotly.express as px
```

```python
# ok, group by city and NAICS, and let's get a count for each
nonprofits_by_type=df_mini.groupby(['CITY','NAICS']).size().reset_index(name='count')
```

```python
# make a stacked bar chart
fig = px.bar(nonprofits_by_type,
           x='CITY',
           y='count',
           color='PRIMARY NAICS DESCRIPTION', # this creates the "stack",
           labels={'CITY':'City in Los Angeles','NAICS':'Number of Nonprofits','PRIMARY NAICS DESCRIPTION':'Type of Industry'},
           title = 'Nonprofits in LA'
      )
fig.show()
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
# create a list of lat/lon pair for UCLA
ucla = [34.0721237,-118.4440685]
m = folium.Map(location=ucla, zoom_start=10)

marker_cluster = MarkerCluster().add_to(m)

for index,row in df.iterrows():
    folium.Marker(
        location=[row.LAT,row.LON],
        popup=row.NAICS,
        icon=folium.Icon(color="red", icon="ok-sign"),
    ).add_to(marker_cluster)
```
